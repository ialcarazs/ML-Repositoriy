# DISEÑO DEL PROYECTO

Este conjunto de datos de nivel principiante tiene 403 filas y 6 columnas. El conjunto de datos se ha dividido en entrenamiento y prueba (entrenamiento: 258, prueba: 145). 

Es un conjunto de datos reales sobre el estado de conocimiento de los estudiantes sobre el tema de Máquinas Eléctricas de CC. 

Este conjunto de datos se recomienda para aprender y practicar sus habilidades en análisis exploratorio de datos , visualización de datos y técnicas de clasificación y clustering . Utilizar diferentes técnicas de aprendizaje supervisadas y no supervisadas .

- Level: Beginner
- Recommended Use: Classification/Clustering
- Domain: Education/Web





## OBJETIVO

Predecir el nivel de conocimiento del estudiante

## ENTIDADES Y DATOS

Información de atributos:

- STG (El grado de tiempo de estudio para materiales de objetos de meta),
- CG (El grado de número de repetición del usuario para los materiales del objeto objetivo)
- STR (El grado de tiempo de estudio del usuario para objetos relacionados con el objeto objetivo)
- LPR (El rendimiento del examen del usuario para objetos relacionados con el objeto objetivo)
- PEG (El rendimiento del examen del usuario para objetos objetivo)
- UNS (El nivel de conocimiento del usuario)
