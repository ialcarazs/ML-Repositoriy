# DISEÑO DEL PROYECTO

Este conjunto de datos de nivel principiante tiene 414 filas y 7 columnas. Proporciona el conjunto de datos históricos del mercado de valoraciones inmobiliarias que se recopilan de Sindian Dist., New Taipei City, Taiwán. Este conjunto de datos se recomienda para aprender y practicar sus habilidades en análisis de datos exploratorios , visualización de datos y técnicas de modelado de regresión

- Nivel: Principiante 
- Uso recomendado: Modelos de regresión 
- Dominio: Inmobiliario

## OBJETIVO

Predecir el precio de la casa
