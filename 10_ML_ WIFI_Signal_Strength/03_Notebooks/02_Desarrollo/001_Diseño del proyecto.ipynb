# Can you estimate the location from WIFI Signal Strength

# DISEÑO DEL PROYECTO

Este conjunto de datos de nivel principiante tiene 2000 filas y 8 columnas. El conjunto de datos contiene la intensidad de la señal wifi observada desde 7 dispositivos wifi en un teléfono inteligente recopilados en un espacio interior. Los datos podrían usarse para estimar la ubicación en una de las cuatro habitaciones.

- Recommended Use: Classification Models
- Domain: Mobile/Location

## OBJETIVO

Predecir la ubicación a partir de la intensidad de la señal wifi.

Este conjunto de datos se recomienda para aprender y practicar sus habilidades en análisis de datos exploratorios , visualización de datos y técnicas de modelado de clasificación . Siéntase libre de explorar el conjunto de datos con múltiples técnicas de aprendizaje supervisadas y no supervisadas .
