## IMPORTACIÓN DE PAQUETES

```python
import numpy as np
import pandas as pd
```

## CREAR LOS DATASETS INICIALES

### CARGAR LOS DATOS

```python
datos = pd.read_csv('../../02_Datos/01_Originales/wifi_localization.txt', sep= ',')
datos
```

### EXTRAER Y RESERVAR EL DATASET DE VALIDACIÓN

```python
val = datos.sample(frac = 0.3).sort_index()
val.head()
```

```python
nombre_val = 'validacion.csv'
ruta = '../../02_Datos/02_Validacion/' + nombre_val

val.to_csv(ruta)
```

### EXTRAER Y RESERVAR EL DATASET DE TRABAJO

```python
trabajo = datos.iloc[~datos.index.isin(val.index)]
trabajo.head()
```

```python
nombre_trabajo = 'trabajo.csv'
ruta = '../../02_Datos/03_Trabajo/' + nombre_trabajo

trabajo.to_csv(ruta)
```
