# Can you predict the fuel-efficieny of a car?

# DISEÑO DEL PROYECTO

## OBJETIVO

- **Level:** Intermediate 
- **Recommended Use:** Regression Models
- **Domain:** Automobiles

Este conjunto de datos de nivel intermedio tiene 398 filas y 9 columnas y proporciona kilometraje, caballos de fuerza, año del modelo y otras especificaciones técnicas para automóviles. Este conjunto de datos se recomienda para aprender y practicar sus habilidades en análisis exploratorio de datos , visualización de datos y técnicas de modelado de regresión . Siéntase libre de explorar el conjunto de datos con múltiples técnicas de aprendizaje supervisadas y no supervisadas.
