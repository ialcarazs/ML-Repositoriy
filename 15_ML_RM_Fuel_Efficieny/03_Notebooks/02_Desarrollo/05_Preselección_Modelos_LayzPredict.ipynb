# 5 - PRESELECCIÓN DEL MODELO LAZY PREDICT

## IMPORTACIÓN DE PAQUETES

```python
import numpy as np 
import pandas as pd

from lazypredict.Supervised import LazyRegressor
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_pickle('../../02_Datos/03_Trabajo/df_tablon.pickle')
df.head()
```

## SEPARAR PREDICTORAS Y TARGET

```python
x = df.drop(columns= 'mpg').copy()
y = df.mpg.copy()
```

```python
x.head()
```

```python
y.value_counts()
```

## SEPARAR DATASET DE ENTRENAMIENTO Y VALIDACIÓN

```python
train_x, val_x, train_y, val_y = train_test_split(x, y, test_size= 0.3, random_state= 42)
```

## MODELIZAR SEGÚN LAZYPREDICT

```python
reg = LazyRegressor(random_state= 42, verbose=0)
models, predictions = reg.fit(train_x, val_x, train_y, val_y)
```

```python
models.sort_values(by = 'RMSE')
```
